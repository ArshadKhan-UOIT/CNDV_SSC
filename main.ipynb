```python
#Is there a  corelation between the number of times an incident is reassigned and how long it takes to restore

#Number of times an incident is reassigned during the life of the ticket.
csvFile3 = pd.read_csv('SSC dataset\INCIDENT_OWNER_HISTORY.csv');
csvFile3 = pd.DataFrame(csvFile3)

grouped = csvFile3.groupby("ticket_nmbr", as_index=False)["assigned_group"].count()
grouped = pd.DataFrame(grouped)

grouped2 = csvFile3.groupby("ticket_nmbr", as_index=False)['TIME_IN_STATUS_BY_OWNER_HRS'].sum()
grouped2 = pd.DataFrame(grouped2)
grouped2 = grouped2.dropna()


result = grouped.merge(grouped2, left_on='ticket_nmbr', right_on='ticket_nmbr')
result = result.groupby("assigned_group", as_index=False)['TIME_IN_STATUS_BY_OWNER_HRS'].mean()
result = pd.DataFrame(result)
result = result.dropna()

result
```
